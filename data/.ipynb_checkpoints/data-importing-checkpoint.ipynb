```python
import pandas as pd
```

```python
# read each csv file into a pandas dataframe
df1 = pd.read_csv('2021-05.csv')
df2 = pd.read_csv('2021-06.csv')
df3 = pd.read_csv('2021-07.csv')

# concatenate the dataframes together
df_journeys = pd.concat([df1, df2, df3])
```

```python
df_journeys
```

```python
df_journeys.isnull().sum()
```

We can drop null distances and also drop journeys that are less than 10 meters in distance or 10 seconds in time

```python
df_journeys = df_journeys[df_journeys['Covered distance (m)'] >= 10]
df_journeys = df_journeys[df_journeys['Duration (sec.)'] >= 10]
```

```python
df_journeys.isnull().sum()
```

```python
df_stations = pd.read_csv('asemat.csv')
df_stations
```

It seems that Helsinki or Helsingfors is missing from columns Kaupunki and Stad we can replace the space by the relevant value

```python
df_stations.loc[df_stations['Kaupunki'] == ' ', 'Kaupunki'] = 'Helsinki'
df_stations.loc[df_stations['Stad'] == ' ', 'Stad'] = 'Helsingfors'

```

```python
df_stations
```

I noticed during development that the CSV file containing stations is missing some stations from the trips CSVs or the other way around, for sake of simplicity I will drop rows of journerys that have departure or return ids that are not in stations table.

```python
stations_id = df_stations['ID'].unique()
```

```python
df_journeys = df_journeys[df_journeys['Departure station id'].isin(stations_id)]
df_journeys = df_journeys[df_journeys['Return station id'].isin(stations_id)]
```

```python
# save to a new csv file
# df_journeys.to_csv('journeys.csv', index=False)
df_stations.to_csv('stations.csv', index=False)
```

It seems like concatenating all trips to one file was not a good idea so we will clean each dataset on its own and import them separately.

```python
df1.info()
```

```python
df1 = df1[df1['Departure station id'].isin(stations_id)]
df1 = df1[df1['Return station id'].isin(stations_id)]
df1 = df1[df1['Covered distance (m)'] >= 10]
df1 = df1[df1['Duration (sec.)'] >= 10]
```

```python
df2 = df2[df2['Departure station id'].isin(stations_id)]
df2 = df2[df2['Return station id'].isin(stations_id)]
df2 = df2[df2['Covered distance (m)'] >= 10]
df2 = df2[df2['Duration (sec.)'] >= 10]
```

```python
df3 = df3[df3['Departure station id'].isin(stations_id)]
df3 = df3[df3['Return station id'].isin(stations_id)]
df3 = df3[df3['Covered distance (m)'] >= 10]
df3 = df3[df3['Duration (sec.)'] >= 10]
```

```python
df1.to_csv('df1.csv', index=False)
df2.to_csv('df2.csv', index=False)
df3.to_csv('df3.csv', index=False)
```
